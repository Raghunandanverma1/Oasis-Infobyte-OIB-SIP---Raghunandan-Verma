**Now , lets see , if any outlier is present in the dataset or not**

```python
x1 = df['Price per Unit']
x2 = df['Total Amount']
x = [x1,x2]
plt.boxplot(x,showmeans = True , tick_labels = ["Price per Unit","Total Unit"])
plt.show()
```

**Therefore we can see that we have no outliers in our datset**

```python
y = df['Quantity']
plt.boxplot(y,showmeans=True, tick_labels=['Quantity'])
plt.show()
```
